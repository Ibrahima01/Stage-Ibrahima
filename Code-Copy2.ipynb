```python
import matplotlib.pyplot as plt
import dask.dataframe as dd
import pandas as pd
```

```python
from sklearn.impute import KNNImputer
import numpy as np
import hashlib
```

```python
df__33SNPs = pd.read_csv('./Sample/00.csv')
```

```python
df__33SNPs.shape
```

```python
# Remplacer les valeurs "NA", "NC", "NN" par des valeurs manquantes NaN
df__33SNPs.replace(['NA', 'NC', 'NN'], np.nan, inplace=True)
```

```python
# Parcourir chaque ligne du DataFrame
for index, row in df__33SNPs.iterrows():

    # Trouver le mot ayant le plus grande occurrence dans la ligne courante
    max_count = 0
    max_word = None
    word_counts = {}
    for word in row:
        if pd.isna(word):
            continue
        if word not in word_counts:
            word_counts[word] = 0
        word_counts[word] += 1
        if word_counts[word] > max_count:
            max_count = word_counts[word]
            max_word = word

    # Remplacer chaque valeur nulle par le mot ayant le plus grande occurrence dans la ligne courante
    for col in df__33SNPs.columns:
        if pd.isna(row[col]):
            df__33SNPs.at[index, col] = max_word

df__33SNPs.to_csv("Imputed_Data.csv")
```
